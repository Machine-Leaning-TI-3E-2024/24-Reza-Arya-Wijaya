```python
# Cek Nilai SSE
print(f'Nilai SSE: {cl_kmeans.inertia_}')
```

**Penjelasan:**

Kode tersebut bertujuan untuk menampilkan nilai Sum of Squared Errors (SSE) dari model K-Means yang telah dilatih. SSE adalah ukuran total variasi dalam cluster, yang dihitung sebagai jumlah kuadrat jarak antara setiap data point dengan centroid clusternya.
* **print(f'Nilai SSE: {cl_kmeans.inertia_}')**: Mencetak nilai SSE ke layar.
  * **cl_kmeans.inertia_:** Atribut ini menyimpan nilai SSE dari model K-Means yang telah dilatih.
  * **f-string**: Digunakan untuk memformat string dan memasukkan nilai SSE ke dalam string yang akan dicetak.

```python
# Implementasi Metode Elbow

# List nilai SSE
sse = []

# Cari k terbaik dari 1-10
K = range(1,10)

# Cek nilai SSE setiap k
for k in K:
 kmeanModel = KMeans(n_clusters=k)
 kmeanModel.fit(X)
 sse.append(kmeanModel.inertia_)


# Plotting the distortions
plt.figure(figsize=(8,4))
plt.plot(K, sse, "bx-")
plt.xlabel("k")
plt.ylabel("SSE")
plt.title("Metode Elbow untuk Mengetahui Jumlah k Terbaik")
plt.show()
```

**Penjelasan:**

Kode tersebut mengimplementasikan metode elbow untuk menentukan jumlah cluster (k) yang optimal dalam algoritma K-Means. Metode elbow bekerja dengan menjalankan K-Means untuk berbagai nilai k dan menghitung SSE untuk setiap nilai k. Kemudian, nilai SSE diplot terhadap nilai k, dan titik "elbow" (siku) pada plot menunjukkan jumlah cluster yang optimal.
* **sse = []**: Inisialisasi list kosong bernama sse untuk menyimpan nilai SSE untuk setiap nilai k.
* **K = range(1, 10)**: Membuat list K yang berisi angka 1 hingga 9, yang merepresentasikan jumlah cluster yang akan diuji.
* **for k in K:**: Loop untuk menjalankan K-Means untuk setiap nilai k dalam list K.
* **kmeanModel = KMeans(n_clusters=k)**: Membuat objek K-Means dengan jumlah cluster k.
* **kmeanModel.fit(X)**: Melatih model K-Means dengan data fitur X.
* **sse.append(kmeanModel.inertia_)**: Menambahkan nilai SSE dari model yang telah dilatih ke dalam list sse.
* **plt.figure(figsize=(8, 4))**: Membuat figure untuk plot dengan ukuran 8x4.
* **plt.plot(K, sse, "bx-")**: Membuat plot garis dengan nilai k pada sumbu x dan nilai SSE pada sumbu y. Marker 'bx-' menunjukkan marker biru berbentuk 'x' dengan garis penghubung.
* **plt.xlabel("k")**: Memberi label "k" pada sumbu x.
* **plt.ylabel("SSE")**: Memberi label "SSE" pada sumbu y.
* **plt.title("Metode Elbow untuk Mengetahui Jumlah k Terbaik")**: Memberi judul pada plot.
* **plt.show()**: Menampilkan plot.

**Algoritma Expectation-Maximization**

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]

    while True:
        # 2a. input label center yang baru
        labels = pairwise_distances_argmin(X, centers)

        # 2b. update center dari titik baru
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])

        # 2c. cek konvergensi
        if np.all(centers == new_centers):
            break
        centers = new_centers

    return centers, labels

centers, labels = find_clusters(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis');
```

**Penjelasan:**

Kode tersebut mengimplementasikan algoritma K-Means dan Expectation-Maximization (EM) untuk melakukan clustering pada data, khususnya data blobs yang dibuat secara sintetis. Kode ini bertujuan untuk mengelompokkan data ke dalam cluster berdasarkan kemiripan fitur-fiturnya, dan kemudian memvisualisasikan hasil clustering tersebut.
* **from sklearn.metrics import pairwise_distances_argmin**: Mengimpor fungsi pairwise_distances_argmin dari modul sklearn.metrics, yang digunakan untuk menghitung jarak antara setiap pasangan data point dan mengembalikan indeks data point terdekat.
* **def find_clusters(X, n_clusters, rseed=2):**: Mendefinisikan fungsi find_clusters yang menerima data fitur X, jumlah cluster n_clusters, dan seed untuk generator angka acak rseed sebagai input.
* **rng = np.random.RandomState(rseed)**: Membuat objek RandomState dari numpy dengan seed yang ditentukan, yang akan digunakan untuk menghasilkan angka acak.
* **i = rng.permutation(X.shape[0])[:n_clusters]**: Mengacak indeks data point dan memilih n_clusters indeks pertama sebagai indeks centroid awal.
* **centers = X[i]**: Mengambil data point yang sesuai dengan indeks yang dipilih sebagai centroid awal.
* **while True:**: Memulai loop yang akan berjalan hingga kondisi konvergensi terpenuhi.
* **labels = pairwise_distances_argmin(X, centers)**: Menghitung jarak antara setiap data point dan centroid, dan menetapkan label cluster berdasarkan centroid terdekat.
* new_centers = np.array([X[labels == i].mean(0) for i in range(n_clusters)])**:
  * Baris ini menghitung pusat cluster baru (new_centers) berdasarkan data point yang telah ditetapkan ke dalam cluster tertentu.
  * **X[labels == i]** memilih data point yang memiliki label cluster sama dengan i.
  * **.mean(0)** menghitung rata-rata dari data point tersebut untuk setiap fitur (dimensi), menghasilkan koordinat pusat cluster baru.
  * **np.array** menggabungkan pusat-pusat cluster baru ini menjadi sebuah array.
* **if np.all(centers == new_centers):**:
  * Baris ini memeriksa apakah pusat cluster baru (new_centers) sama dengan pusat cluster sebelumnya (centers).
  * **np.all** mengembalikan True jika semua elemen dalam perbandingan tersebut bernilai True (artinya, semua koordinat pusat cluster sama).
  * Jika pusat cluster baru sama dengan yang sebelumnya, berarti algoritma telah konvergen, dan loop akan dihentikan.
* **break**: Perintah ini menghentikan loop while jika kondisi konvergensi terpenuhi.
* **centers = new_centers**: Jika pusat cluster belum konvergen, pusat cluster sebelumnya (centers) akan diperbarui dengan pusat cluster baru (new_centers), dan loop akan berlanjut ke iterasi berikutnya.
* **return centers, labels**: Setelah loop selesai (baik karena konvergensi atau mencapai jumlah iterasi maksimum), fungsi akan mengembalikan pusat cluster akhir (centers) dan label cluster untuk setiap data point (labels).
* **centers, labels = find_clusters(X, 4)**:
  * Baris ini memanggil fungsi find_clusters dengan data fitur X dan jumlah cluster yang diinginkan (4).
  * Fungsi **find_clusters** akan melakukan proses clustering dan mengembalikan pusat cluster akhir (centers) dan label cluster untuk setiap data point (labels).
  * Nilai yang dikembalikan oleh fungsi tersebut kemudian disimpan dalam variabel centers dan labels.
* **plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')**:
  * Baris ini membuat scatter plot untuk memvisualisasikan hasil clustering.
  * **X[:, 0] dan X[:, 1]** digunakan sebagai koordinat x dan y untuk setiap data point.
  * **c=labels** memberikan warna pada data point berdasarkan label cluster yang telah ditetapkan.

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

**Penjelasan:**

```python
# Fungsi untuk menghitung jarak antara vektor input dan bobot SOM
def calculate_distance(input_vector, weights):
    return np.linalg.norm(input_vector - weights, axis=2)
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama calculate_distance yang bertujuan untuk menghitung jarak antara vektor input (piksel warna dari gambar) dan bobot dari neuron-neuron dalam model Self-Organizing Map (SOM).
* **def calculate_distance(input_vector, weights):**:
  * Mendeklarasikan sebuah fungsi bernama calculate_distance yang menerima dua argumen:
    * **input_vector**: Vektor yang merepresentasikan piksel warna dari gambar input.
    * **weights**: Array yang berisi bobot dari neuron-neuron SOM.
* **return np.linalg.norm(input_vector - weights, axis=2)**:
  * Fungsi ini mengembalikan sebuah array yang berisi jarak antara vektor input dan bobot SOM.
  * **np.linalg.norm()** digunakan untuk menghitung norma (magnitude) dari vektor, yang dalam hal ini merepresentasikan jarak Euclidean.
  * **input_vector - weights**: Menghitung selisih antara vektor input dan bobot SOM.
  * **axis=2**: Menentukan bahwa perhitungan norma dilakukan sepanjang sumbu terakhir (sumbu saluran warna).

```python
# Fungsi untuk menemukan indeks unit pemenang (unit dengan bobot terdekat)
def find_winner_unit_in_image(input_vector, weights):
    distances = calculate_distance(input_vector, weights)
    return np.unravel_index(np.argmin(distances), distances.shape)
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama find_winner_unit_in_image yang bertujuan untuk menemukan indeks dari neuron pemenang (Best Matching Unit/BMU) dalam model Self-Organizing Map (SOM) untuk suatu vektor input (piksel warna dari gambar).
* **def find_winner_unit_in_image(input_vector, weights):**:
  * Mendeklarasikan sebuah fungsi bernama find_winner_unit_in_image yang menerima dua argumen:
    * **input_vector**: Vektor yang merepresentasikan piksel warna dari gambar input.
    * **weights**: Array yang berisi bobot dari neuron-neuron SOM.
* **distances = calculate_distance(input_vector, weights)**:
  * Memanggil fungsi calculate_distance (yang telah didefinisikan sebelumnya) untuk menghitung jarak antara vektor input dan bobot dari setiap neuron dalam SOM.
  * Hasil perhitungan jarak disimpan dalam variabel distances.
* **return np.unravel_index(np.argmin(distances), distances.shape)**:
  * Menemukan indeks dari neuron yang memiliki jarak terkecil (yaitu, BMU) menggunakan np.argmin(distances).
  * **np.unravel_index()** digunakan untuk mengubah indeks linier menjadi indeks multidimensi yang sesuai dengan dimensi grid SOM.
  * Fungsi ini mengembalikan indeks BMU dalam bentuk tuple (baris, kolom).

```python
# Fungsi untuk memperbarui bobot SOM
def update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius):
    distances = np.linalg.norm(np.indices(weights.shape[:2]).T - np.array(winner).reshape(1, -1), axis=2)
    influence = np.exp(-distances / (2 * neighborhood_radius**2))
    weights += learning_rate * influence[:, :, np.newaxis] * (input_vector - weights)
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama update_weights yang bertujuan untuk memperbarui bobot dari neuron-neuron dalam model Self-Organizing Map (SOM) selama proses pelatihan. Pembaruan bobot dilakukan berdasarkan neuron pemenang (BMU) dan radius tetangga.
* **def update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius):**:
  * Mendeklarasikan sebuah fungsi bernama update_weights yang menerima lima argumen:
    * **input_vector**: Vektor yang merepresentasikan piksel warna dari gambar input.
    * **weights**: Array yang berisi bobot dari neuron-neuron SOM.
    * **winner**: Tuple yang berisi indeks dari neuron pemenang (BMU).
    * **learning_rate**: Laju pembelajaran, yang mengontrol seberapa besar bobot diperbarui.
    * **neighborhood_radius**: Radius tetangga, yang mengontrol seberapa luas area neuron yang diperbarui.
* **distances = np.linalg.norm(np.indices(weights.shape[:2]).T - np.array(winner).reshape(1, -1), axis=2)**:
  * Menghitung jarak antara setiap neuron dalam grid SOM dan BMU.
  * **np.indices()** digunakan untuk membuat array indeks untuk grid SOM.
  * **np.linalg.norm()** digunakan untuk menghitung jarak Euclidean.
* **influence = np.exp(-distances / (2 * neighborhood_radius**2))**:
   * Menghitung pengaruh (influence) dari BMU terhadap setiap neuron berdasarkan jarak dan radius tetangga.
  * Pengaruh ini dihitung menggunakan fungsi Gaussian.
* **weights += learning_rate * influence[:, :, np.newaxis] * (input_vector - weights)**: Memperbarui bobot dari setiap neuron berdasarkan pengaruh BMU, laju pembelajaran, dan selisih antara vektor input dan bobot neuron.
---

```python
# Fungsi untuk melatih SOM
def train_som(image, num_epochs, initial_learning_rate, initial_neighborhood_radius):
    input_shape = image.shape
    som_shape = (10, 10, input_shape[2])  # Ukuran SOM sesuai dengan jumlah saluran warna
    weights = initialize_weights(input_shape, som_shape)

    for epoch in range(num_epochs):
        # Update parameter pembelajaran dan radius tetangga
        learning_rate = initial_learning_rate * np.exp(-epoch / num_epochs)
        neighborhood_radius = initial_neighborhood_radius * np.exp(-epoch / num_epochs)
        # Pemrosesan SOM
        for i in range(input_shape[0]):
            for j in range(input_shape[1]):
                input_vector = image[i, j, :]
                winner = find_winner_unit_in_image(input_vector, weights)
                update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius)

    return weights
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama train_som yang bertujuan untuk melatih model Self-Organizing Map (SOM) menggunakan gambar input. Fungsi ini melakukan iterasi melalui setiap piksel dalam gambar, menemukan neuron pemenang (BMU), dan memperbarui bobot neuron untuk mempelajari pola warna dalam gambar.
* **def train_som(image, num_epochs, initial_learning_rate, initial_neighborhood_radius):**:
  * Mendeklarasikan sebuah fungsi bernama train_som yang menerima empat argumen:
    * **image**: Gambar input yang akan digunakan untuk melatih SOM.
    * **num_epochs**: Jumlah epoch pelatihan.
    * **initial_learning_rate**: Laju pembelajaran awal.
    * **initial_neighborhood_radius**: Radius tetangga awal.
* **input_shape = image.shape**: Mendapatkan dimensi gambar input.
* **som_shape = (10, 10, input_shape[2])**: Menentukan ukuran grid SOM dan jumlah saluran warna.
* **weights = initialize_weights(input_shape, som_shape)**: Menginisialisasi bobot SOM secara acak.
* **for epoch in range(num_epochs):**: Melakukan iterasi melalui setiap epoch pelatihan.
* **learning_rate = initial_learning_rate * (1 - epoch / num_epochs)**: Memperbarui laju pembelajaran.
* **neighborhood_radius = initial_neighborhood_radius * (1 - epoch / num_epochs)**: Memperbarui radius tetangga.
* **for i in range(input_shape[0]):**: Melakukan iterasi melalui setiap baris piksel dalam gambar.
* **for j in range(input_shape[1]):**: Melakukan iterasi melalui setiap kolom piksel dalam gambar.
* **input_vector = image[i, j, :]**: Mendapatkan vektor input (piksel warna).
* **winner = find_winner_unit_in_image(input_vector, weights)**: Menemukan BMU.
* **update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius)**: Memperbarui bobot neuron.
* **return weights**: Mengembalikan bobot SOM yang telah dilatih.
* **for epoch in range(num_epochs):**:
Loop ini mengiterasi melalui setiap epoch selama proses pelatihan SOM. num_epochs adalah jumlah total epoch yang ditentukan saat memanggil fungsi train_som.
* **learning_rate = initial_learning_rate * np.exp(-epoch / num_epochs)**:
  * Memperbarui nilai learning_rate (laju pembelajaran) pada setiap epoch.
  * Laju pembelajaran dikurangi secara eksponensial seiring bertambahnya epoch.
  * Ini dilakukan untuk memungkinkan SOM melakukan penyesuaian yang lebih halus pada bobot neuron seiring berjalannya pelatihan.
* **neighborhood_radius = initial_neighborhood_radius * np.exp(-epoch / num_epochs)**:
  * Memperbarui nilai neighborhood_radius (radius tetangga) pada setiap epoch.
  * Radius tetangga juga dikurangi secara eksponensial seiring bertambahnya epoch.
  * Ini dilakukan untuk mempersempit area pengaruh neuron pemenang (BMU) selama pelatihan, sehingga SOM dapat mempelajari pola yang lebih detail.
* **for i in range(input_shape[0]):**: Loop luar ini mengiterasi melalui setiap baris piksel dalam gambar input.
* **for j in range(input_shape[1]):**: Loop dalam ini mengiterasi melalui setiap kolom piksel dalam gambar input.
* **input_vector = image[i, j, :]**:
  * Mengambil nilai piksel warna pada posisi (i, j) dari gambar input dan menyimpannya dalam variabel input_vector.
  * **input_vector** akan digunakan sebagai input untuk proses SOM selanjutnya, seperti menemukan BMU dan memperbarui bobot.
* **winner = find_winner_unit_in_image(input_vector, weights)**:
  * Memanggil fungsi find_winner_unit_in_image yang telah didefinisikan sebelumnya.
  * Fungsi ini akan mencari neuron pemenang (BMU) untuk input_vector (piksel warna saat ini) dalam grid SOM.
  * Indeks BMU disimpan dalam variabel winner.
* **update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius)**:
  * Memanggil fungsi update_weights, juga telah didefinisikan sebelumnya.
  * Fungsi ini akan memperbarui bobot neuron-neuron dalam SOM berdasarkan BMU yang ditemukan, laju pembelajaran (learning_rate), dan radius tetangga (neighborhood_radius).
  * Tujuannya adalah untuk membuat bobot neuron-neuron di sekitar BMU menjadi lebih mirip dengan input_vector (warna piksel saat ini), sehingga SOM dapat mempelajari pola warna dalam gambar.
* **return weights**: Setelah loop pelatihan selesai (semua piksel dalam gambar telah diproses untuk setiap epoch), fungsi train_som mengembalikan weights, yaitu bobot SOM yang telah dilatih dan diperbarui.

**Hitung Jarak Vektor Input dan Bobot SOM**

```python
# Fungsi untuk menghitung jarak antara vektor input dan bobot SOM
def calculate_distance(input_vector, weights):
    return np.linalg.norm(input_vector - weights, axis=2)
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama calculate_distance yang bertujuan untuk menghitung jarak antara vektor input (piksel warna dari gambar) dan bobot dari neuron-neuron dalam model Self-Organizing Map (SOM).
* **def calculate_distance(input_vector, weights):**:
  * Mendeklarasikan sebuah fungsi bernama calculate_distance yang menerima dua argumen:
    * **input_vector**: Vektor yang merepresentasikan piksel warna dari gambar input.
    * **weights**: Array yang berisi bobot dari neuron-neuron SOM.
* **return np.linalg.norm(input_vector - weights, axis=2)**:
  * Fungsi ini mengembalikan sebuah array yang berisi jarak antara vektor input dan bobot SOM.
  * **np.linalg.norm()** digunakan untuk menghitung norma (magnitude) dari vektor, yang dalam hal ini merepresentasikan jarak Euclidean.
  * **input_vector - weights**: Menghitung selisih antara vektor input dan bobot SOM.
  * **axis=2**: Menentukan bahwa perhitungan norma dilakukan sepanjang sumbu terakhir (sumbu saluran warna).

**Unit Dengan Bobot Terdekat**

```python
# Fungsi untuk menemukan indeks unit pemenang (unit dengan bobot terdekat)
def find_winner_unit_in_image(input_vector, weights):
    distances = calculate_distance(input_vector, weights)
    return np.unravel_index(np.argmin(distances), distances.shape)
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama find_winner_unit_in_image yang bertujuan untuk menemukan indeks dari neuron pemenang (Best Matching Unit/BMU) dalam model Self-Organizing Map (SOM) untuk suatu vektor input (piksel warna dari gambar).
* **def find_winner_unit_in_image(input_vector, weights):**:
  * Mendeklarasikan sebuah fungsi bernama find_winner_unit_in_image yang menerima dua argumen:
    * **input_vector**: Vektor yang merepresentasikan piksel warna dari gambar input.
    * **weights**: Array yang berisi bobot dari neuron-neuron SOM.
* **distances = calculate_distance(input_vector, weights)**:
  * Memanggil fungsi calculate_distance (yang telah didefinisikan sebelumnya) untuk menghitung jarak antara vektor input dan bobot dari setiap neuron dalam SOM.
  * Hasil perhitungan jarak disimpan dalam variabel distances.
* **return np.unravel_index(np.argmin(distances), distances.shape)**:
  * Menemukan indeks dari neuron yang memiliki jarak terkecil (yaitu, BMU) menggunakan np.argmin(distances).
  * **np.unravel_index()** digunakan untuk mengubah indeks linier menjadi indeks multidimensi yang sesuai dengan dimensi grid SOM.
  * Fungsi ini mengembalikan indeks BMU dalam bentuk tuple (baris, kolom).

**Perbarui Bobot SOM**

```python
# Fungsi untuk memperbarui bobot SOM
def update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius):
    distances = np.linalg.norm(np.indices(weights.shape[:2]).T - np.array(winner).reshape(1, -1), axis=2)
    influence = np.exp(-distances / (2 * neighborhood_radius**2))
    weights += learning_rate * influence[:, :, np.newaxis] * (input_vector - weights)
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama update_weights yang bertujuan untuk memperbarui bobot dari neuron-neuron dalam model Self-Organizing Map (SOM) selama proses pelatihan. Pembaruan bobot dilakukan berdasarkan neuron pemenang (BMU) dan radius tetangga.
* **def update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius):**:
  * Mendeklarasikan sebuah fungsi bernama update_weights yang menerima lima argumen:
    * **input_vector**: Vektor yang merepresentasikan piksel warna dari gambar input.
    * **weights**: Array yang berisi bobot dari neuron-neuron SOM.
    * **winner**: Tuple yang berisi indeks dari neuron pemenang (BMU).
    * **learning_rate**: Laju pembelajaran, yang mengontrol seberapa besar bobot diperbarui.
    * **neighborhood_radius**: Radius tetangga, yang mengontrol seberapa luas area neuron yang diperbarui.
* **distances = np.linalg.norm(np.indices(weights.shape[:2]).T - np.array(winner).reshape(1, -1), axis=2)**:
  * Menghitung jarak antara setiap neuron dalam grid SOM dan BMU.
  * **np.indices()** digunakan untuk membuat array indeks untuk grid SOM.
  * **np.linalg.norm()** digunakan untuk menghitung jarak Euclidean.
* **influence = np.exp(-distances / (2 * neighborhood_radius**2))**:
   * Menghitung pengaruh (influence) dari BMU terhadap setiap neuron berdasarkan jarak dan radius tetangga.
  * Pengaruh ini dihitung menggunakan fungsi Gaussian.
* **weights += learning_rate * influence[:, :, np.newaxis] * (input_vector - weights)**: Memperbarui bobot dari setiap neuron berdasarkan pengaruh BMU, laju pembelajaran, dan selisih antara vektor input dan bobot neuron.

**Latih SOM**

```python
# Fungsi untuk melatih SOM
def train_som(image, num_epochs, initial_learning_rate, initial_neighborhood_radius):
    input_shape = image.shape
    som_shape = (10, 10, input_shape[2])  # Ukuran SOM sesuai dengan jumlah saluran warna
    weights = initialize_weights(input_shape, som_shape)

    for epoch in range(num_epochs):
        # Update parameter pembelajaran dan radius tetangga
        learning_rate = initial_learning_rate * np.exp(-epoch / num_epochs)
        neighborhood_radius = initial_neighborhood_radius * np.exp(-epoch / num_epochs)
        # Pemrosesan SOM
        for i in range(input_shape[0]):
            for j in range(input_shape[1]):
                input_vector = image[i, j, :]
                winner = find_winner_unit_in_image(input_vector, weights)
                update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius)

    return weights
```

**Penjelasan:**

Kode tersebut mendefinisikan sebuah fungsi bernama train_som yang bertujuan untuk melatih model Self-Organizing Map (SOM) menggunakan gambar input. Fungsi ini melakukan iterasi melalui setiap piksel dalam gambar, menemukan neuron pemenang (BMU), dan memperbarui bobot neuron untuk mempelajari pola warna dalam gambar.
* **def train_som(image, num_epochs, initial_learning_rate, initial_neighborhood_radius):**:
  * Mendeklarasikan sebuah fungsi bernama train_som yang menerima empat argumen:
    * **image**: Gambar input yang akan digunakan untuk melatih SOM.
    * **num_epochs**: Jumlah epoch pelatihan.
    * **initial_learning_rate**: Laju pembelajaran awal.
    * **initial_neighborhood_radius**: Radius tetangga awal.
* **input_shape = image.shape**: Mendapatkan dimensi gambar input.
* **som_shape = (10, 10, input_shape[2])**: Menentukan ukuran grid SOM dan jumlah saluran warna.
* **weights = initialize_weights(input_shape, som_shape)**: Menginisialisasi bobot SOM secara acak.
* **for epoch in range(num_epochs):**: Melakukan iterasi melalui setiap epoch pelatihan.
* **learning_rate = initial_learning_rate * (1 - epoch / num_epochs)**: Memperbarui laju pembelajaran.
* **neighborhood_radius = initial_neighborhood_radius * (1 - epoch / num_epochs)**: Memperbarui radius tetangga.
* **for i in range(input_shape[0]):**: Melakukan iterasi melalui setiap baris piksel dalam gambar.
* **for j in range(input_shape[1]):**: Melakukan iterasi melalui setiap kolom piksel dalam gambar.
* **input_vector = image[i, j, :]**: Mendapatkan vektor input (piksel warna).
* **winner = find_winner_unit_in_image(input_vector, weights)**: Menemukan BMU.
* **update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius)**: Memperbarui bobot neuron.
* **return weights**: Mengembalikan bobot SOM yang telah dilatih.
* **for epoch in range(num_epochs):**:
Loop ini mengiterasi melalui setiap epoch selama proses pelatihan SOM. num_epochs adalah jumlah total epoch yang ditentukan saat memanggil fungsi train_som.
* **learning_rate = initial_learning_rate * np.exp(-epoch / num_epochs)**:
  * Memperbarui nilai learning_rate (laju pembelajaran) pada setiap epoch.
  * Laju pembelajaran dikurangi secara eksponensial seiring bertambahnya epoch.
  * Ini dilakukan untuk memungkinkan SOM melakukan penyesuaian yang lebih halus pada bobot neuron seiring berjalannya pelatihan.
* **neighborhood_radius = initial_neighborhood_radius * np.exp(-epoch / num_epochs)**:
  * Memperbarui nilai neighborhood_radius (radius tetangga) pada setiap epoch.
  * Radius tetangga juga dikurangi secara eksponensial seiring bertambahnya epoch.
  * Ini dilakukan untuk mempersempit area pengaruh neuron pemenang (BMU) selama pelatihan, sehingga SOM dapat mempelajari pola yang lebih detail.
* **for i in range(input_shape[0]):**: Loop luar ini mengiterasi melalui setiap baris piksel dalam gambar input.
* **for j in range(input_shape[1]):**: Loop dalam ini mengiterasi melalui setiap kolom piksel dalam gambar input.
* **input_vector = image[i, j, :]**:
  * Mengambil nilai piksel warna pada posisi (i, j) dari gambar input dan menyimpannya dalam variabel input_vector.
  * **input_vector** akan digunakan sebagai input untuk proses SOM selanjutnya, seperti menemukan BMU dan memperbarui bobot.
* **winner = find_winner_unit_in_image(input_vector, weights)**:
  * Memanggil fungsi find_winner_unit_in_image yang telah didefinisikan sebelumnya.
  * Fungsi ini akan mencari neuron pemenang (BMU) untuk input_vector (piksel warna saat ini) dalam grid SOM.
  * Indeks BMU disimpan dalam variabel winner.
* **update_weights(input_vector, weights, winner, learning_rate, neighborhood_radius)**:
  * Memanggil fungsi update_weights, juga telah didefinisikan sebelumnya.
  * Fungsi ini akan memperbarui bobot neuron-neuron dalam SOM berdasarkan BMU yang ditemukan, laju pembelajaran (learning_rate), dan radius tetangga (neighborhood_radius).
  * Tujuannya adalah untuk membuat bobot neuron-neuron di sekitar BMU menjadi lebih mirip dengan input_vector (warna piksel saat ini), sehingga SOM dapat mempelajari pola warna dalam gambar.
* **return weights**: Setelah loop pelatihan selesai (semua piksel dalam gambar telah diproses untuk setiap epoch), fungsi train_som mengembalikan weights, yaitu bobot SOM yang telah dilatih dan diperbarui.
